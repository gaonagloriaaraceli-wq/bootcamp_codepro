```python
#definimos la funcion 
def encontrar_palabra(palabra_adivinar,palabra_usuario):
    cantidad_letras=5
    #creamos una lista para almacenar las letras correctas 
    letras_correctas=[]
    
    #usamos un for para pasar por cada posicion de la palabra
    for posicion in range (cantidad_letras):
        #si la palabra es exactamente igual
        palabras_iguales=palabra_usuario[posicion]==palabra_adivinar[posicion]
        #si hay una letra existente dentro de la palabra del usuario 
        letra_existente=palabra_usuario[posicion]in palabra_adivinar[posicion]

        #usaremos la condicion if para verificar y agregar a la lista si se cumple

        if palabras_iguales:
            letras_correctas.append("["+palabra_usuario[posicion]+"]")

        elif letra_existente:
            letras_correctas.append("("+palabra_usuario[posicion]+")")
        else:
            letras_correctas.append(palabra_usuario[posicion])
            
    return letras_correctas
resultado= encontrar_palabra("gatos","datos")
print(resultado)



```
